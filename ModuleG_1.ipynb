Riemanns Integral and Trapezoid Rule

by Trapezoid Rule,



```python
dx=.1
x=np.linspace(0,10, int(10/dx))

f=np.cos(x)*x


print('Our starting function')
plt.plot(x,f, label='cos(x)*x')

#code for calculating an approximation by Riemanns integral
sum=0
F=np.zeros_like(f)
for i in range(len(f)):
  sum+=f[i]*dx
  F[i]=sum

print('Simple numerically integrated function')
plt.plot(x,F, label='Riemanns integral(cos(x)*x)')


#code for calculating an approximation by Trapezoid Rule integral
sum=0
TrapezoidF=np.zeros(len(f))
for i in range(len(f)-1):
  sum+=(f[i]+f[1+i])/2*dx
  TrapezoidF[i+1]=sum

plt.plot(x,TrapezoidF, label='Trapezoid Rule integral(cos(x)*x)')


#code for calculating real value
realAnswer =  np.cos(x)+x*np.sin(x)
plt.plot(x,realAnswer, label='real integral(cos(x)*x)')


plt.legend()
print('Error = %.2f' % np.mean( (F-realAnswer)**2)**.5)
```
