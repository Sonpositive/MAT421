1. Interpolation

Interpolation in statistics refers to the process of estimating the value of a function for a point that lies within the range of a set of known data points.

```python
import matplotlib.pyplot as plt


plt.style.use('seaborn-poster')


x=[0,1,2,3,4]
y=[0,2.5,4.6,2.1,0.5]

plt.figure(figsize= (10,8))

plt.plot(x,y,'bo');

plt.show()
```
