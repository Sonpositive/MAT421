**1.Based-N and Binary**

Example of binary encoding of numbers

```python
print(10,  '(decmal)=', bin(10), '(binary)\n')
print(15,  '(decmal)=', bin(15), '(binary)\n')
```

Operation of binary numbers

```python
x = 5
y = 6

# Convert to binary strings
binary_x = bin(x)[2:]  # '101'
binary_y = bin(y)[2:]  # '110'

# Convert binary strings back to integers and multiply
result = int(binary_x, 2) * int(binary_y, 2)

print(f'{binary_x} * {binary_y} = {bin(result)}')
```



**3.Round-off Errors**


Round-off error is the difference between an approximation of a number used in computation and its true value

```python
#round-off error
1.82 - 0.815 == 1.005
```

```python
#calculation
1.82 - 0.815
```

Though the numbers cannot be made closer to their intended exact values, the round function can be useful for
post-rounding so that results with inexact values become comparable to one another:

```python
round(1.82 - 0.815, 3) == 1.005
```



Accumulation of round-off error happens when we doing a sequence of calculation on an initial input with round-off input, then error can be accumulated.


```python
# If we only run it once
2 + 1/7 - 1/7

```

```python
def add_and_subtract(iterations):
  result = 2

  for i in range(iterations):
    result += 1/7
  for i in range(iterations):
    result -= 1/7
  return result
```

```python
# If we do this 100 times
add_and_subtract(100)
```

```python
# If we do this 1000 times
add_and_subtract(1000)
```

```python
# If we do this 10000 times
add_and_subtract(10000)
```
